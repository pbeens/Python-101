# How to Get Input from the Computer


We use the `input()` command and assign it to a **variable**, which in this case is named "`name`". *Variables will be discussed in more detail below.*

We can also include an optional quote (`'What is your name? '`, below) which will be printed to the screen.

- [[Extra help]](https://drive.google.com/file/d/1kIdKv-axlRRFggZ1YYKIOGuzpLsxikBU/view?usp=sharing)




````python
name = input('What is your name? ')
````

Notice that I have single-quotes around the prompt (which is technically called a string). Python also accepts double-quotes. This is a matter of personal style. 

We can print the variable, which we have suitable called `name`. Notice there aren't any quotes around the variable name. If we put quotes around 'name' it would literally print the word name instead of our name.

````python
print(name)
````

At this stage it is important to note that anything inputted this way is assumed to be text-based and not numeric. We will deal with the inputting of numbers later.

# An Introduction to Variables

A variable is like a container for storing data values. 

Variables can store basic items like numbers, letters, words, and even groups of items. 

image.png

A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume).

Rules for Python variables:

* A variable name must start with a letter or the underscore character
* A variable name cannot start with a number
* A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
* Variable names are case-sensitive (age, Age and AGE are three different variables)

````python
# Legal variable names:
myvar = "John"
my_var = "John"
_my_var = "John"
myVar = "John"
MYVAR = "John"
myvar2 = "John"

# Illegal variable names:
2myvar = "John"
my-var = "John"
my var = "John"
````
